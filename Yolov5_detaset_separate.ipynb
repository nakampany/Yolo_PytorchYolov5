````python
trainfiles = glob.glob(os.path.join(dir_path, 'train/images/*.jpg'))
valfiles = glob.glob(os.path.join(dir_path, 'val/images/*.jpg'))

for i,f in enumerate(trainfiles):
  os.rename(f, os.path.join(dir_path, 'train/images/train_' + '{0:04d}'.format(i) + '.jpg'))
  
for i,f in enumerate(valfiles):
  os.rename(f, os.path.join(dir_path, 'val/images/val_' + '{0:04d}'.format(i) + '.jpg'))
````

````python
trainfiles = glob.glob(os.path.join(dir_path, 'train/labels/*.txt'))
valfiles = glob.glob(os.path.join(dir_path, 'val/labels/*.txt'))

for i,f in enumerate(trainfiles):
  os.rename(f, os.path.join(dir_path, 'train/labels/train_' + '{0:04d}'.format(i) + '.txt'))
  
for i,f in enumerate(valfiles):
  os.rename(f, os.path.join(dir_path, 'val/labels/val_' + '{0:04d}'.format(i) + '.txt'))
````
